```python
import re

with open("menu_category.html") as f:
    lines = f.readlines()

lines = [line.strip() for line in lines]
```

```python
category_to_id_dict = {}

for line in lines:
    if not line.startswith('<li class='):
        continue
    try:
        cat_id = re.findall("menu-category-(\d+)", line)[0]
        cat_name = re.findall('class=".*">(.+)</a>', line)[0].replace("&amp;", "&")
        print(f"Detect {cat_id} - {cat_name}")
        category_to_id_dict[cat_name] = cat_id
    except Exception as e:
        print(f"Error at {line}; message: {str(e)}")

print(f"Detect {len(category_to_id_dict.keys())} category in total")
```

```python
import json

with open("category.json", "w") as f:
    f.write(json.dumps(category_to_id_dict, indent=4))
```
